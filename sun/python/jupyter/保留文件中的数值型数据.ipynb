```python
import pandas as pd
import numpy as np
df = pd.read_csv('D:/data/sun/final/location_all.csv')
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError: 
        pass
    try:
        import unicodedata  
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
    return False
```

```python
def str_indexlookup(df):
    list1 = []
    for i in range(df.shape[0]):
        if i%10000 == 0:
            print('i=' , i)
        if is_number(df['pre_night_longitude'][i]) and is_number(df['pre_night_latitude'][i]):
            pass
        else:
            list1.append(i)
    return list1
```

```python
df_index = [i for i in range(df.shape[0])]
str_index = str_indexlookup(df)
num_list = list(set(df_index)-set(str_index))
```
