```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
```

```python
df = pd.read_csv('D:/data/sun/location_wrong1.csv')
#df.dropna(subset=['east', 'west','north','south'])
df.isnull().sum()
df = df[df['north'] < 90]
df = df[df['south'] < 90]
df = df[df['east'] < 180]
df = df[df['west'] < 180]
df = df[df['north_1'] < 90]
df = df[df['south_1'] < 90]
df = df[df['east_1'] < 180]
df = df[df['west_1'] < 180]
ew_list = []
ns_list = []
ew_list1 = []
ns_list1 = []
```

```python
def area_match(df):
    row_index = np.arange(0,df.shape[0],1)
    global ew_list 
    global ns_list
    ew_list =[]
    ns_list =[]
    for i in row_index:
        east = df.iloc[i:i+1,1:2].get_values()
        west = df.iloc[i:i+1,2:3].get_values()
        north = df.iloc[i:i+1,3:4].get_values()
        south = df.iloc[i:i+1,4:5].get_values()
        ew = geodesic((north,east), (north,west))
        ns = geodesic((north,east), (south,east))
        ew_list.append(ew)
        ns_list.append(ns)
    return ew_list,ns_list

def area_match1(df):
    row_index = np.arange(0,df.shape[0],1)
    global ew_list1
    global ns_list1
    ew_list1 =[]
    ns_list1 =[]
    for i in row_index:
        east1 = df.iloc[i:i+1,7:8].get_values()
        west1 = df.iloc[i:i+1,8:9].get_values()
        north1 = df.iloc[i:i+1,9:10].get_values()
        south1 = df.iloc[i:i+1,10:11].get_values()
        ew1 = geodesic((north1,east1), (north1,west1))
        ns1 = geodesic((north1,east1), (south1,east1))
        ew_list1.append(ew1)
        ns_list1.append(ns1)
    return ew_list1,ns_list1
```

```python
area_match(df)
area_match1(df)
```

```python
df['ew'] = ew_list
df['ns'] = ns_list
df['ew1'] = ew_list1
df['ns1'] = ns_list1
```

```python
df.to_csv('D:/data/sun/location_wrong11.csv')
```
