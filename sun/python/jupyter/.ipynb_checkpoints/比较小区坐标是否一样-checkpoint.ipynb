```python
import pandas as pd
import numpy as np
```

```python
xqdf = pd.read_csv('D:/data/sun/calxqusercount/label_use_result.csv',dtype={
    'west_longitude':np.longdouble,
    'east_longitude':np.longdouble,
    'north_latitude':np.longdouble,
    'south_latitude':np.longdouble
})
mx_xqdf = pd.read_csv('D:/data/sun/calxqusercount/label_use_result_model.csv',dtype={
    'west_longitude_mx':np.longdouble,
    'east_longitude_mx':np.longdouble,
    'north_latitude_mx':np.longdouble,
    'south_latitude_mx':np.longdouble
})
```

```python
df = pd.merge(xqdf,mx_xqdf,on='xq_code',how = 'inner')
num = 0
```

```python
for i in range(0,df.shape[0]):
    print("当前计算第",i,"个小区")
    if (df['west_longitude'][i] == df['west_longitude_mx'][i]) and (df['east_longitude'][i] == df['east_longitude_mx'][i]) and (df['north_latitude'][i] == df['north_latitude_mx'][i]) and (df['south_latitude'][i] == df['south_latitude_mx'][i]):
        num += 1
num
```
