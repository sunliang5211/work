```python
import pandas as pd
import numpy as np
```

```python
#df = pd.read_csv('D:/data/sun/scanstr/location_all_num.csv')
df = pd.read_csv('D:/data/sun/scanstr/person_wz.csv')
```

```python
xqdf = pd.read_csv('D:/data/sun/calxqusercount_0819/xqwz.csv',dtype={
    'west_longitude':np.longdouble,
    'east_longitude':np.longdouble,
    'north_latitude':np.longdouble,
    'south_latitude':np.longdouble
})
```

```python
xqdf['usr_count'] = 0
for i in range(0,xqdf.shape[0]):
    print("当前计算第",i,"个小区")
    test_df = df[(df.pre_night_longitude>xqdf['west_longitude'][i])&
            (df.pre_night_longitude<xqdf['east_longitude'][i])&
            (df.pre_night_latitude<xqdf['north_latitude'][i])&
            (df.pre_night_latitude>xqdf['south_latitude'][i])]
    xqdf['usr_count'][i] = test_df.shape[0]
```

```python
xqdf['usr_count'].sum()
xqdf.to_csv('D:/data/sun/calxqusercount_0819/xqwz_usercount1.csv')
```
