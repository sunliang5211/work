```python
# -*- coding:utf-8 -*-
#导入相关包
import os,sys
import time
import pandas as pd
import numpy as np
import lightgbm as lgb
test_savepath = './'
model_savepath = './'
res_savepath = './'
```

```python
#df = pd.read_csv(test_savepath+sys.argv[1],low_memory=False)
df = pd.read_csv(test_savepath+'06allfea_train_B232080.csv',low_memory=False)
test = df.iloc[:,1:df.shape[1]-1]
```

```python
#bst = lgb.Booster(model_file=model_savepath+sys.argv[3])
bst = lgb.Booster(model_file=model_savepath+'lgb_B232084')
ypred = bst.predict(test, num_iteration=bst.best_iteration)
```

```python
df_pre = pd.DataFrame(ypred)
df_pre['label'] = df['label']
df_pre.columns = ['pre','label']
```

```python
df_pre_prelabel1 = df_pre.loc[df_pre['pre'] > df_pre['pre'].quantile(0.99)]
df_pre_testlabel1 = df_pre.loc[df_pre['label'] == 1]
```

```python
precision = df_pre_prelabel1['label'].sum()/df_pre_prelabel1.shape[0]
recall = df_pre_testlabel1.loc[df_pre_testlabel1['pre'] > df_pre['pre'].quantile(0.99)].shape[0]/df_pre_testlabel1.shape[0]
f1 = 2*precision*recall/(precision+recall)
```

```python
print('precision:',"%.4f%%" % (precision * 100))
print('recall:',"%.4f%%" % (recall * 100))
print('f1',"%.4f%%" % (f1 * 100))
```

```python
df_res = df_pre.loc[df_pre['pre'] > df_pre['pre'].quantile(0.99)]
df_res.to_csv(res_savepath+sys.argv[2])
```
